```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('student-scores.csv')
```

```python
dataset.isnull().sum()
```

```python
data=dataset.drop(columns=['id','first_name','last_name','email'],axis=1)
```

```python
data
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
encoded=ohe.fit_transform(data[['gender','part_time_job','extracurricular_activities','career_aspiration']])
data=data.drop(columns=['gender','part_time_job','extracurricular_activities','career_aspiration'],axis=1)
encoded_df=pd.DataFrame(encoded,columns=ohe.get_feature_names_out())
df=pd.concat([data,encoded_df],axis=1)
```

```python
df
```

```python
df['absence_days'].max()
```

```python
df.info()
```

```python
y=df['math_score']
```

```python
y
```

```python
x=df.drop(columns=['math_score','history_score','physics_score','chemistry_score','biology_score','english_score','geography_score'],axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor,plot_tree
model=DecisionTreeRegressor(random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={'max_depth':[None,7,8,9,10,15,20,35],'min_samples_split':[2,3,8,9,5,20,25,35],'min_samples_leaf':[5,1,8,15,20,32,200,100,150,330]}
grid=GridSearchCV(model,params,scoring='neg_mean_squared_error')
```

```python
grid.fit(x_train,y_train)
```

```python
print(grid.best_params_)
```

```python
print(grid.best_score_)
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
mse=mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
rmse=np.sqrt(mse)
print(rmse)
```
